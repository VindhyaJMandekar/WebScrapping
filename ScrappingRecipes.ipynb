```python
import requests
from bs4 import BeautifulSoup as bs
import time
import urllib.parse
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0'}

base='https://www.ambitiouskitchen.com/recipe-index/?_sft_mealtypes=dinner'
```

```python
content=requests.get(base,headers=headers)
soup=bs(content.content,'html.parser')

#soup = BeautifulSoup(text)

data = soup.findAll('div',attrs={'class':'search-filter-results'})
recipes=[]
for div in data:
    links = div.findAll('a')
    for a in links:
        recipes.append(a['href'])
recipes = list(set(recipes))
recipes
```

```python
#recipe_df = pd.DataFrame(columns=['RecipeNo','RecipeName', 'IngredientName', 'IngredientAmount','IngredientUnit'])
i=1
j=1
recipe_list = []
for link in recipes:
    recipeCont=requests.get(link,headers=headers)
    soup=bs(recipeCont.content,'html.parser')
    recipe_title_div=soup.select_one('div.dmc-col-1 h2')
    RecipeNo=i
    if(recipe_title_div == None):
        continue
    RecipeName=recipe_title_div.text
    lines = soup.find_all('li', attrs={'class':'wprm-recipe-ingredient'})
    for line in lines:
        recipe_line={}
        amount = line.find('span',attrs={'class':'wprm-recipe-ingredient-amount'})
        if(amount == None):
            amount = '0'
        else:
            amount=amount.getText()
        unit = line.find('span',attrs={'class':'wprm-recipe-ingredient-unit'})
        if(unit == None):
            unit = ''
        else:
            unit=unit.getText()
        name = line.find('span',attrs={'class':'wprm-recipe-ingredient-name'})
        if(name == None):
            name = ''
        else:
            name=name.getText()
        recipe_line={'RecipeNo':RecipeNo,'RecipeName':RecipeName,'IngredientName':name,'IngredientAmount':amount,'IngredientUnit':unit}
        recipe_list.append(recipe_line)
        j=j+1
    i=i+1
    time.sleep(5)
recipe_df=pd.DataFrame(recipe_list)
recipe_df
```
